色補正

```python
import numpy as np
from PIL import Image

def apply_color_matrix(image_path, color_matrix):
    # 画像を開く
    img = Image.open(image_path)

    # 画像をRGBモードに変換
    img = img.convert('RGB')

    # 画像をNumPy配列に変換
    img_array = np.array(img, dtype=np.float32)

    # 色補正行列を適用
    # ここで色補正行列を画像のRGB値に適用します
    img_array = img_array @ color_matrix

    # 結果を0-255の範囲にクリッピングして整数型に戻す
    img_array = np.clip(img_array, 0, 255).astype(np.uint8)

    # NumPy配列をPIL画像に変換
    corrected_img = Image.fromarray(img_array)
    corrected_img = corrected_img.convert('RGB')

    return corrected_img


data_path = 'imgs\demo_re\corrected'
for b in range(2):
    if b == 0:
        path = 'nobacklight'
    else:
        path = 'backlight'
    for i in range(9):
        print(i)
        color_matrix = np.load(f'{data_path}/{path}/{i:03d}.npz')['color_matrix_lstsq']
        img_corrected = apply_color_matrix(f'{data_path}/{path}/{i:03d}.jpg', color_matrix)
        img_corrected.save(f'{data_path}/{path}/{i:03d}.jpg')
```

解像度を下げる

```python
from PIL import Image

def reduce_image_resolution(input_path, output_path, w, h):
    """
    Reduce the resolution of an image by a specified scale factor.

    Args:
        input_path (str): Path to the input image.
        output_path (str): Path to save the output image with reduced resolution.
        scale_factor (float): Factor by which to reduce the resolution (e.g., 0.5 for half size).
    """
    # Open the input image
    with Image.open(input_path) as img:
        # Calculate the new dimensions
        new_width = w
        new_height = h
        
        # Resize the image
        reduced_img = img.resize((new_width, new_height))
        
        # Save the reduced resolution image
        reduced_img.save(output_path)
        print(f"Image saved at {output_path} with resolution {new_width}x{new_height}")

# Example usage

data_path = 'imgs\leaf9\corrected'
for b in range(2):
    if b == 0:
        path = 'nobacklight'
    else:
        path = 'backlight'
    for i in range(128):
        print(i)
        reduce_image_resolution(f'{data_path}/{path}/{i:03d}.jpg', f'{data_path}/{path}/{i:03d}.jpg', 900, 600)
```
